```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Replace with your model repo name
model_name = "2oofs/llama3python"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# If you have a GPU, move the model to it for faster inference
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
prompt = "write a function to create a class"

inputs = tokenizer(prompt, return_tensors="pt").to(device)

# Generate output tokens (adjust max_new_tokens as needed)
outputs = model.generate(**inputs, max_new_tokens=100)

# Decode generated tokens to text
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(generated_text)
```
